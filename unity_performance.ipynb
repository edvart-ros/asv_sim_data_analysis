```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

data_empty = pd.read_csv("data/FpsEmptyWorld.csv")
data_ocean = pd.read_csv("data/FpsDefaultOcean.csv")
data_dynamics = pd.read_csv("data/FpsWamvKernerBuoyancyDynamics.csv")
data_camera = pd.read_csv("data/FpsWamvCamera.csv")
data_all_sensors = pd.read_csv("data/FpsWamvAllSensors.csv")

average_empty = np.mean(data_empty[" fps "])
average_ocean = np.mean(data_ocean[" fps "])
average_single_boat = np.mean(data_dynamics[" fps "])
average_sensor_camera = np.mean(data_camera[" fps "])
average_all_sensors = np.mean(data_all_sensors[" fps "])
# Set a style
plt.style.use('seaborn-v0_8-deep')

# Increase the default figure size
plt.figure(figsize=(10, 6))

# Increase default font size
mpl.rcParams.update({'font.size': 14})

# Plot with increased line width and add markers
plt.plot(data_empty["Time"], data_empty[" fps "], label =f"Empty World", linewidth=2, marker='o')
plt.plot(data_ocean["Time"], data_ocean[" fps "], label = "Ocean only", linewidth=2, marker='s')
plt.plot(data_dynamics["Time"], data_dynamics[" fps "], label = "With single boat", linewidth=2, marker='^')
plt.plot(data_camera["Time"], data_camera[" fps "], label = "With sensor camera", linewidth=2, marker='>')
plt.plot(data_all_sensors["Time"], data_all_sensors[" fps "], label = "With all sensors", linewidth=2, marker='x')

# Add grid, labels, and title
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Frames per Second (fps)')

# Set y-limits and add a tight layout
plt.ylim((0, 450))
plt.tight_layout()

# Optimize legend location and make it semi-transparent
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=5, fancybox=True).get_frame().set_alpha(0.5)

# Show the plot
plt.savefig("results/performance_plot.pdf", bbox_inches='tight')
```
